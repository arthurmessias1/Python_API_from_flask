```python
import pandas as pd
from flask import Flask, jsonify

app = Flask(__name__)

# Homepage padrão da API "Site"
@app.route('/')
def homepage():
    return 'A API está no ar'

# Pagina somente do total de vendas da tabela utilizada na API
@app.route('/dados')
def totalvendas():
    tabela = pd.read_csv('api.csv')
    total_vendas = tabela ['Vendas'].sum()
    resposta = {'total_vendas' : total_vendas}
    return jsonify(resposta)

# Página da tabela utilizada na API completa convertida em dicionário
@app.route('/tabela')
def tabelaapi():
    dict_from_csv = pd.read_csv('api.csv', header=0, index_col=0, squeeze=True).to_dict()
    return jsonify(dict_from_csv)

# Criação de um dicionário aleatório para teste
@app.route('/tabelateste')
def dados():
    tabela2 = [
    {'id': 0,
     'title': 'A Fire Upon the Deep',
     'author': 'Vernor Vinge',
     'first_sentence': 'The coldsleep itself was dreamless.',
     'year_published': '1992'},
    {'id': 1,
     'title': 'The Ones Who Walk Away From Omelas',
     'author': 'Ursula K. Le Guin',
     'first_sentence': 'With a clamor of bells that set the swallows soaring, the Festival of Summer came to the city Omelas, bright-towered by the sea.',
     'published': '1973'},
    {'id': 2,
     'title': 'Dhalgren',
     'author': 'Samuel R. Delany',
     'first_sentence': 'to wound the autumnal city.',
     'published': '1975'}
]
   
    return jsonify(tabela2)

app.run()
```
